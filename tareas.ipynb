# Voy a describir las variables en este espacio y luego haré una tabla para que sea más fácil observarlas
# Alcohol: Cuantitativa continua. Representa el porcentaje de alcohol en el vino, puede tomar cualquier valor dentro de un rango.
# Ácido málico: Cuantitativa continua. Es una concentración de ácido málico, también un valor continuo.
# Cenizas: Cuantitativa continua. Es la proporción de cenizas en el vino, un valor continuo.
# Alcalinidad de cenizas: Cuantitativa continua. Medida de la alcalinidad en las cenizas, valor continuo.
# Magnesio: Cuantitativa discreta. Es el contenido de magnesio, un valor entero.
# Fenoles totales: Cuantitativa continua. Es la cantidad total de fenoles presentes en el vino.
# Flavonoides: Cuantitativa continua. Es la cantidad de flavonoides en el vino.
# Fenoles no flavonoides: Cuantitativa continua. Similar a los anteriores, mide fenoles específicos.
# Proantocianidinas: Cuantitativa continua. Otro compuesto químico en el vino, valor continuo.
# Intensidad de color: Cuantitativa continua. Mide la intensidad del color del vino.
# Matiz: Cuantitativa continua. Representa el matiz del color del vino.
# OD280/OD315 de vinos diluidos: Cuantitativa continua. Medida de absorbancia óptica.
# Prolina: Cuantitativa continua. Concentración de prolina en el vino, valor continuo.

```python
import pandas as pd
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
import seaborn as sns

wine = fetch_openml('wine', version=1, as_frame=True)
wine_frame = pd.DataFrame(wine.data)

wine_frame.columns = ['Alcohol', 'Ácido málico', 'Cenizas', 
                      'Alcalinidad de cenizas', 'Magnesio',
                      'Fenoles totales', 'Flavonoides', 
                      'Fenoles no flavonoides', 'Proantocianidinas',
                      'Intensidad de color', 'Matiz', 
                      'OD280/OD315 de vinos diluidos', 'Prolina']

tipos_variables = wine_frame.dtypes
print("Tipos de variables:\n", tipos_variables)

estadisticas = wine_frame.describe()
print("\nEstadísticas descriptivas:\n", estadisticas)

plt.figure(figsize=(10, 5))
sns.histplot(wine_frame['Alcohol'], bins=30, kde=True)
plt.title('Distribución del Alcohol en el Vino')
plt.xlabel('Alcohol')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(data=wine_frame, x='Alcohol', y='Ácido málico', hue='Ácido málico', palette='viridis')
plt.title('Gráfico de Dispersión: Alcohol vs Ácido Málico')
plt.xlabel('Alcohol')
plt.ylabel('Ácido Málico')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.boxplot(y='Fenoles totales', data=wine_frame)
plt.title('Diagrama de Caja: Fenoles Totales')
plt.ylabel('Fenoles Totales')
plt.grid(True)
plt.show()
```
